```python
# Import essential libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')  # built-in to seaborn
```

```python
iris.shape           # shows rows and columns, e.g., (150, 5)
iris.head()          # first 5 rows
iris.info()          # column types and non-null counts
iris.describe()      # summary stats (mean, min, max)
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
plt.title('Sepal vs Petal Length by Species')
```

```python
iris.hist(figsize=(10, 8), bins=15)
plt.suptitle('Feature Distributions')
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=iris, orient="h")
plt.title('Box Plot for Numeric Features')
```

### 🧠 Data Visualization Explanations

#### 📌 Why Scatter Plots Help Show Relationships
Scatter plots are used to explore the relationship between two continuous variables. Each dot represents one data point. If the dots form a pattern (like a line or a curve), it indicates a correlation between the variables — positive, negative, or none.  
In the Iris dataset, scatter plots (e.g., petal length vs petal width) help us see how different species cluster and how strongly their features are related.

---

#### 📌 What Histograms Reveal
Histograms display the distribution of a single feature by dividing data into intervals (bins). They help identify:
- Whether data is **normally distributed**
- If it's **skewed** to the left or right
- If there are **multiple peaks** in the data  
For example, a histogram of sepal width in the Iris dataset can show how frequently certain width ranges occur and if the data is concentrated or spread out.

---

#### 📌 How Box Plots Show Range and Outliers
Box plots summarize the distribution of a variable by showing:
- **Median** (center line)
- **Interquartile range (IQR)** (the box itself)
- **Whiskers** (min and max within 1.5×IQR)
- **Outliers** (points beyond the whiskers)  
They are great for spotting outliers and comparing distributions across categories (e.g., petal length across species in the Iris dataset).

---


### 📘 Summary of Insights

- **Sepal and petal lengths tend to increase together**, showing a positive correlation between these features.
- **Setosa flowers** are clearly distinguishable as they have **smaller petal length and width** compared to Versicolor and Virginica.
- **Versicolor and Virginica** overlap more in their feature values but can still be separated using petal dimensions.
- **Histograms** revealed that some features are more **normally distributed**, while others (like petal length) are **skewed** due to class differences.
- **Box plots** helped identify **outliers** and compare feature ranges across different species.

This simple data exploration gives a strong understanding of how species differ and which features are most useful for classification.
